```python
# Import Modules
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
file_name = "hzs.jpeg"
my_img = Image.open(file_name)
img_array = np.array(my_img)
```

```python
img_array.shape
```

```python
plt.imshow(img_array)
```

```python
# Reshape array along axis 0 and add count for unique colors
color, count = np.unique(img_array.reshape(-1, 3), axis=0, return_counts=True)
```

```python
# Add count as a column and sort in descending order
result = np.column_stack((color, count))
```

```python
# Use argsort to find the indices and then print the color array with those indices
count_list = result[:,3]
ascending_indeces = np.argsort(count_list)
indeces = ascending_indeces[::-1][:10]
```

```python
for n in range(10):
    print(color[indeces[n]])
    plt.imshow([[(color[indeces[n]])]])
    plt.show()
```
